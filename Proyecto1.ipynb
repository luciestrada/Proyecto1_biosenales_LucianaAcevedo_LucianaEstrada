Filtrado


El funcionamiento de los filtros se caracteriza matemáticamente por su respuesta en la frecuencia, la cual se obtiene mediante la transformada de Fourier. Esta respuesta está compuesta por la magnitud y la fase, parámetros que definen las propiedades del sistema y describen exactamente cómo el filtro altera cada componente frecuencial de la señal de entrada para generar una salida filtrada.

En cuanto a la respuesta en frecuencia, el efecto del filtro sobre la señal temporal varía drásticamente según su diseño. Los filtros de paso bajo atenúan las altas frecuencias, lo que resulta en un suavizado de la señal. Por el contrario, los filtros de paso alto eliminan los componentes de baja frecuencia y el desplazamiento de corriente continua (DC). Finalmente, los filtros de rechazo de banda (Notch), usados para eliminar el ruido eléctrico (50/60 Hz), requieren precaución por su tendencia a generar artefactos oscilatorios; por ello, se recomiendan alternativas como la regresión en el dominio temporal (Widmann et al., 2015).

Los filtros IIR operan de forma funcional y recursiva, lo que les permite lograr cortes abruptos con alta eficiencia computacional; sin embargo, presentan fase no lineal y riesgo de inestabilidad por errores de redondeo.En comparación, los filtros FIR garantizan estabilidad absoluta y fase lineal, pero requieren órdenes muy elevados para lograr transiciones de banda comparables, lo que aumenta significativamente el costo computacional al procesar los múltiples canales y archivos de la base de datos. Para este proyecto, donde se deben procesar señales de muchos sujetos en varias condiciones, la eficiencia computacional del filtro IIR Butterworth es una ventaja práctica importante. Además, dado que el análisis se basa en densidad espectral de potencia (PSD) y no en la morfología exacta de eventos temporales, la distorsión de fase no lineal del IIR no representa un problema crítico, ya que la PSD es insensible a la fase de la señal.

Widmann, A., Schröger, E., & Maess, B. (2015). Digital filter design for electrophysiological data--a practical approach. Journal of neuroscience methods, 250, 34–46. https://doi.org/10.1016/j.jneumeth.2014.08.002

La razón principal del filtrado, en general, es atenuar el ruido en las grabaciones, preservando la señal (de interés).

Se utilizan filtros digitales (generalmente filtros IIR o FIR) para atenuar las frecuencias que no aportan información. El filtro Pasa-Banda (8 Hz a 30 Hz) aísla los ritmos Mu (8-12 Hz) y Beta (13-30 Hz). El filtro Notch (50 Hz o 60 Hz) elimina la interferencia electromagnética de los enchufes y equipos cercanos.

Es importante tener en cuenta que los cambios en el espectro de frecuencia (la atenuación o el retardo de los componentes espectrales) deben provocar cambios en la señal temporal , ya que ambas representaciones están acopladas por la transformada de Fourier . Por lo tanto, se debe tener cuidado al seleccionar y diseñar el filtro; es decir, los parámetros del filtro deben ajustarse para lograr una mejor relación señal-ruido y así estimar los parámetros. Sin embargo, es imposible diseñar filtros que no alteren la señal en absoluto. En cambio, al seleccionar el filtro adecuado, la señal se modifica según los objetivos del investigador y, por lo tanto, aumenta su relación señal-ruido.

Filtro de paso bajo: Se encarga de atenuar las bandas de alta frecuencia. Su efecto principal en el dominio del tiempo es el suavizado de la señal, eliminando las variaciones rápidas.
Filtro de paso alto: Su función es atenuar las bandas de baja frecuencia y eliminar el desplazamiento de corriente continua (DC). Esto fuerza a que la señal filtrada converja rápidamente a una amplitud cero tras una deflexión temporal.
Filtro de paso banda: Consiste teóricamente en la combinación de un filtro paso bajo y uno paso alto, permitiendo el paso de un rango específico de frecuencias. Sin embargo, para evitar distorsiones marcadas en el dominio del tiempo, la literatura sugiere que a menudo es preferible aplicar de forma sucesiva un filtro paso alto pronunciado y un filtro paso bajo más suave, en lugar de un paso banda estricto.
Filtro de rechazo de banda (Notch): Está diseñado para atenuar bandas de frecuencia muy específicas. En electrofisiología, su uso se limita casi exclusivamente a la supresión del ruido de la red eléctrica (50/60 Hz) o de pantallas. No obstante, se desaconseja su uso generalizado debido a que tienden a producir fuertes artefactos en la señal, sugiriéndose en su lugar métodos de limpieza alternativos como la regresión en el dominio del tiempo.



Selección de Canales y Filtrado Espacial

El problema principal de extraer los ritmos mu y beta de la corteza sensoriomotora para el control de un BCI independiente radica en el fenómeno de conducción de volumen. Cuando las neuronas generan actividad eléctrica al imaginar un movimiento, esa corriente tiene que atravesar múltiples capas de tejido de alta resistencia como el líquido cefalorraquídeo, el cráneo y el cuero cabelludo antes de llegar al electrodo. Durante ese trayecto la señal se esparce y se difumina. Como resultado no solo se capta la actividad de la mano, sino también el ruido de fondo proveniente de todo el cerebro (Wolpaw et al., 2002).

Por esta razón es tan importante la extracción precisa de intenciones motoras a partir de señales EEG, enfocando anatómicamente en la corteza motora primaria, como c3,c4 y cz. Sin embargo, debido a este efecto de conducción de volumen, es indispensable aplicar un filtrado espacial.  De acuerdo con el estudio fundamental de McFarland et al. (1997), la aplicación de filtros espaciales como la Referencia Común Promedio (CAR) o las derivaciones Laplacianas mejora significativamente la relación señal-ruido.

McFarland, D. J., McCane, L. M., David, S. V., & Wolpaw, J. R. (1997). Spatial filter selection for EEG-based communication. Electroencephalography and clinical neurophysiology, 103(3), 386–394. https://doi.org/10.1016/s0013-4694(97)00022-2


Wolpaw, J. R., Birbaumer, N., McFarland, D. J., Pfurtscheller, G., & Vaughan, T. M. (2002). Brain–computer interfaces for communication and control. Clinical Neurophysiology, 113(6), 767–791. https://doi.org/10.1016/S1388-2457(02)00057-3

Segmentación por épocas y limpieza
Después del filtrado, el siguiente paso consiste en dividir el registro continuo en segmentos temporales discretos denominados épocas. Cada época corresponde a un intervalo de tiempo alineado con un evento de interés, en este caso los marcadores indican el movimiento de la mano derecha e izquierda y la imaginación de movimiento de la mano derecha, izquierda y en reposo.
La segmentación se realiza definiendo una ventana temporal con un offset de inicio y uno de fin respecto al marcador del evento, lo que permite capturar con precisión las señales motoras correspondientes a cada condición evaluada. La elección del tamaño de la ventana depende del fenómeno que se desea analizar: para capturar los fenómenos de ERD/ERS asociados a los ritmos mu y beta, se recomienda incluir un período de línea base previo al estímulo (típicamente entre −1 y −0.5 s) y un período activo posterior de al menos 2 a 4 segundos, ya que estos ritmos presentan dinámicas lentas que requieren ventanas suficientemente largas para ser resueltos espectralmente. Extracción de características (PSD por banda)

Las señales EEG se adquieren en el dominio del tiempo, por lo que es necesario transformarlas al dominio de la frecuencia para extraer las características espectrales importantes. Para eso empleamos el método de Welch, que estima la densidad espectral de potencias (PSD) dividiendo la señal en segmentos solapados, aplicando una ventana a cada uno y promediando sus periodogramas, lo que reduce la varianza de la estimación respecto a una FFT directa. Este método ha sido confirmado como uno de los más efectivos para estimar características de potencia espectral en señales de imaginación motora (Li et al., 2024). Este procedimiento permite construir representaciones espacio frecuenciales que capturan simultáneamente las características espectrales de la señal y la distribución topológica de los electrodos sobre el cuero cabelludo




Tienes señales en el dominio del tiempo (voltaje vs tiempo). Necesitas pasarlas al dominio de la frecuencia usando el Método de Welch para obtener la Densidad Espectral de Potencia (PSD). Esto te dirá cuánta "energía" hay en la banda Mu y cuánta en la banda Beta. Si hay movimiento/imaginación, la energía debería bajar (ERD - Desincronización).

Análisis estadístico (comparar condiciones)


El objetivo fundamental es determinar si las variaciones en la Densidad Espectral de Potencia (PSD) de los ritmos Mu y Beta constituyen diferencias estadísticamente significativas atribuibles al fenómeno de Desincronización Relacionada a Eventos (ERD). Para ello, se extraen las características frecuenciales de los canales motores (C3, C4, Cz) y se someten a pruebas de contraste de hipótesis, tales como la prueba t de Student (para dos condiciones) o el Análisis de Varianza (ANOVA). Este rigor estadístico permite confirmar que las características suministradas al clasificador representan verdaderos patrones neurofisiológicos subyacentes a la intención motora y no fluctuaciones aleatorias del ruido de fondo.

Comparar las cuatro condiciones experimentales requiere verificar que las variaciones en la PSD de los ritmos mu y beta no sean producto del azar. Con este fin, se extraen los valores de potencia por banda en los canales C3, C4 y Cz, y se aplican pruebas de hipótesis cuya elección depende del cumplimiento de los supuestos estadísticos: si los datos siguen una distribución normal (Shapiro-Wilk) y presentan varianzas homogéneas (Levene), se emplean pruebas paramétricas como ANOVA o t de Student; de lo contrario, se recurre a alternativas no paramétricas como Kruskal-Wallis. De esta forma se puede establecer con rigor si las diferencias espectrales observadas entre movimiento real, imaginación de movimiento y reposo corresponden genuinamente a los fenómenos ERD/ERS propios de la actividad sensoriomotora, y no a fluctuaciones inherentes al ruido de la señal.

# **Plan de análisis**



Se seleccionaron 5 sujetos , y se extrajo la información de los electrodos representativos de la red sensoriomotora con el fin de capturar tanto la ejecución como la imaginación motora. Para ello se incluyen los canales C3, y  C4 como principales, debido a su relación directa con la corteza motora primaria y el control contralateral de las manos. Además, se incorporan FC3 ,FC4, CP3, y CP4, asociados a la corteza premotora que juega un papel en la planificación e imaginación del movimiento. Esta selección permite evaluar tanto la activación focal como los patrones más distribuidos que caracterizan a la imaginación motora.

Preprocesamiento de la señal EEG


El preprocesamiento de la señal EEG se realiza mediante una secuencia de filtrado diseñada para eliminar interferencias externas y mantener las bandas de interés.
- Filtro Pasa-Altas: Se aplica un filtro pasa-altas Butterworth (orden 4) con frecuencia de corte en 1 Hz para eliminar la deriva de baja frecuencia y artefactos lentos asociados al movimiento de electrodos.
- Filtro Pasa-Bajas: Se implementa un filtro pasa-bajas Butterworth (orden 4) con corte en 40 Hz para suprimir componentes de alta frecuencia como el ruido muscular , resultando en un ancho de banda efectivo de 1–40 Hz, ideal para el análisis de ritmos mu y beta.
- Filtro Notch: Finalmente, se aplica un filtro notch a 60 Hz (IIR, orden 2) para remover el ruido de la red eléctrica.


Todo el filtrado se realiza utilizando el método de fase cero (filtfilt) para evitar distorsiones temporales en la señal. Una vez filtrada la señal, se aplica una re-referenciación tipo promedio común (CAR), que consiste en restar el promedio de todos los canales en cada instante de tiempo., para resaltar la actividad local. Posteriormente, la señal se segmenta en ventanas temporales (epochs) de –1 a 4.5 segundos (Duración según la base de datos) , utilizando el intervalo  de –1 a 0 segundos como baseline (Referencia que indica que no se ha realizado la respectiva acción).

Extracción de características

Para evidenciar la diferencia en ritmos cerebrales asociados las cuatro condiciones de estudio (MR-D, MR-I, IM-D , IM-I), se lleva a cabo un análisis compuesto por 3 secciones centradas en identificar individualmente el comportamiento de las diferentes características de los diferentes estudios de la base de datos .


- Análisis por canales y ritmos

En primer lugar se busca identificar los ritmos más discriminantes, y  qué electrodos muestran mayor activación en las diferentes condiciones . Para esto se calcula la densidad espectral de potencia (PSD) mediante el método de Welch, y la transformada rápida de Fourier.  Este análisis compara las bandas Mu vs Beta vs Theta , sin embargo se enfoca en las bandas mu (8–13 Hz) y beta (13–30 Hz), ya que estos son los marcadores estándar de la actividad sensoriomotora. Adicionalmente este análisis evalúa el lado derecho en las dos condiciones (MR, y IM) , en los canales C3, FC3, y CP3, ya que son consistentes con la correspondencia entre los patrones de movimiento de las manos y la corteza cerebral en neurofisiología, además partimos de que la ejecución motora genera una activación localizada,y  la imaginación motora tiende a mostrar una activación más distribuida que alcanza las regiones premotoras.

- Análisis de lateralidad

Para identificar si la condición es de derecha o izquierda , se realiza un análisis que involucra mapas topográficos para visualizar la distribución de la actividad cortical durante cada condición, evidenciando la dominancia contralateral esperada. Asimismo, se calcula el índice de lateralidad definido como LI = (C3 − C4) / (C3 + C4), con el fin de cuantificar la asimetría hemisférica. Se espera que el LI sea negativo para tareas de mano derecha, indicando mayor activación en C3 (hemisferio izquierdo, contralateral), y positivo para tareas de mano izquierda. En este caso se hace un estudio donde se compara MR-D vs MR-I  y IM-D vs IM-I, la misma acción y  en la banda mu porque ….., y en el canal C3 y C4 , proque estos corresponden o se relacionan con C3 a la derecha y el C4 a la izquierda , esto nos permite encontrar qué lado refleja  la dominancia contralateral .

- Análisis según el evento

Para lograr distinguir entre el movimiento real y la imaginería motora, se emplean ventanas deslizantes sobre cada época que permiten observar la evolución de la potencia a lo largo del tiempo, identificando el inicio del ERD y la aparición del post-movement beta rebound (PMBR), caracterizado por un incremento de la potencia beta posterior a la finalización del movimiento. Por ende en este caso se realiza la comparación de MR-D y IM-D, evaluados en la banda mu y canal C3 , y  MR-I y IM-I, evaluados en la banda mu y canal C4, permitiendo identificar características entre la ejecución y la imaginación. Adicionalmente para evaluar  el post movimiento beta-rebound se examina el lado derecho , donde se compara MR y IM , en la banda BETA y en el canal C3, con este fenómeno se logra reconocer la recuperación fisiológica del sistema motor que se da solo tras movimiento real. Este estudio se evalúa cuantitativamente por medio del porcentaje de desincronización relacionada a eventos (ERD%) (ERD% = ((P_activo − P_baseline) / P_baseline) × 100), este indicador permite evaluar la magnitud de la activación cortical durante cada condición, siendo esperable una mayor desincronización (valores más negativos) durante la ejecución del movimiento en comparación con la imaginación motora.

```python
import os
import urllib.request
from google.colab import drive

# ─────────────────────────────────────────
# MONTAR DRIVE Y DESCARGAR DATOS
# ─────────────────────────────────────────
drive.mount('/content/drive')

RUTA_DRIVE = '/content/drive/MyDrive/Proyecto_BCI_2026'
os.makedirs(RUTA_DRIVE, exist_ok=True)

# Solo descargamos los 5 sujetos y los runs que necesitamos
SUJETOS_DESCARGAR = ['sub-001', 'sub-002', 'sub-003', 'sub-004', 'sub-005']
RUNS_DESCARGAR    = range(1, 9)  # runs 1-8
EXTENSIONES       = ['_eeg.set', '_events.tsv', '_channels.tsv', '_eeg.json']
BASE_URL          = "https://openneuro.org/crn/datasets/ds004362/files"

def descargar_archivo(url, destino):
    """Descarga un archivo solo si no existe ya."""
    if os.path.exists(destino) and os.path.getsize(destino) > 0:
        return "ya existe"
    os.makedirs(os.path.dirname(destino), exist_ok=True)
    try:
        urllib.request.urlretrieve(url, destino)
        return "ok"
    except Exception as e:
        return f"error: {e}"

print("Descargando datos...\n")
for sub in SUJETOS_DESCARGAR:
    print(f"  {sub}...")
    for run in RUNS_DESCARGAR:
        for ext in EXTENSIONES:
            nombre  = f"{sub}_task-motion_run-{run}{ext}"
            url     = f"{BASE_URL}/{sub}:eeg:{nombre}"
            destino = os.path.join(RUTA_DRIVE, sub, 'eeg', nombre)
            estado  = descargar_archivo(url, destino)
            if estado not in ("ya existe", "ok"):
                print(f"    {nombre}: {estado}")

    # Verifica que quedaron los runs necesarios
    sets = [f for f in os.listdir(os.path.join(RUTA_DRIVE, sub, 'eeg'))
            if f.endswith('.set')]
    print(f"     {len(sets)} archivos .set descargados")

print("\n Descarga completa.")
```

```python
import mne
import numpy as np
import pandas as pd
import os

# ─────────────────────────────────────────
# CONFIGURACIÓN
# ─────────────────────────────────────────
RUTA_DRIVE = '/content/drive/MyDrive/Proyecto_BCI_2026'
SUJETOS = ['sub-001', 'sub-002', 'sub-005', 'sub-003', 'sub-004']
RUNS_INTERES = [3, 4, 7, 8]  # Task1 manos y Task2 imaginación
CANALES_INTERES = ['C3', 'C4', 'Cz', 'Fc3', 'Fc4', 'Cp3', 'Cp4']

# ─────────────────────────────────────────
# CARGA
# ─────────────────────────────────────────
datos = {} #diccionario vacío

for sub in SUJETOS:
    print(f"\nCargando {sub}...") #Pasa por cada persona la lista
    datos[sub] = {}

    for run_num in RUNS_INTERES:  #Busca específicamente los experimentos de interés (movimiento e imaginación).
        ruta_set = os.path.join(RUTA_DRIVE, sub, 'eeg',
                                f"{sub}_task-motion_run-{run_num}_eeg.set")
        ruta_ev  = os.path.join(RUTA_DRIVE, sub, 'eeg',
                                f"{sub}_task-motion_run-{run_num}_events.tsv")

        if not os.path.exists(ruta_set):
            print(f"  run-{run_num}: no encontrado")
            continue

        raw = mne.io.read_raw_eeglab(ruta_set, preload=True, verbose=False)
        disponibles = [c for c in CANALES_INTERES if c in raw.ch_names]  #De los 64 electrodos que tiene el casco, el código solo se queda con los CANALES_INTERES.
        raw.pick(disponibles)
        df_ev = pd.read_csv(ruta_ev, sep='\t')

        datos[sub][run_num] = {'raw': raw, 'events': df_ev}
        print(f"  run-{run_num}: OK | canales: {raw.ch_names} | eventos: {len(df_ev)}")

print("\nCarga completa.")
```

```python
# Se aplica sobre raw_paso1 (ya tiene pasa-altas)
raw_paso2 = raw_paso1.copy().filter(
    l_freq=None, h_freq=40.0,
    method='iir', iir_params=dict(order=4, ftype='butter'),
    phase='zero', verbose=False
)
raw_paso2.plot(title="3. Pasa-altas + Pasa-bajas 40 Hz → elimina ruido muscular")
```

```python
# Se aplica sobre raw_paso2 (ya tiene ambos filtros)
raw_paso3 = raw_paso2.copy().notch_filter(
    freqs=60.0, method='iir',
    iir_params=dict(order=2, ftype='butter'),
    verbose=False
)
raw_paso3.plot(title="4. + Notch 60 Hz → elimina ruido red eléctrica")
```

```python
# Último paso: CAR. Este raw_paso4 es la señal completamente limpia
raw_paso4 = raw_paso3.copy().set_eeg_reference(
    'average', projection=False, verbose=False
)
raw_paso4.plot(title="5. + CAR → señal final limpia")

# ── Ahora aplicamos este mismo flujo a TODOS los sujetos ──
def preprocesar_raw(raw):
    raw = raw.copy()
    raw.filter(l_freq=1.0, h_freq=None, method='iir',          # paso 1
               iir_params=dict(order=4, ftype='butter'), phase='zero', verbose=False)
    raw.filter(l_freq=None, h_freq=40.0, method='iir',         # paso 2
               iir_params=dict(order=4, ftype='butter'), phase='zero', verbose=False)
    raw.notch_filter(freqs=60.0, method='iir',                  # paso 3
                     iir_params=dict(order=2, ftype='butter'), verbose=False)
    raw.set_eeg_reference('average', projection=False, verbose=False)  # paso 4
    return raw

for sub in SUJETOS:
    for run_num in datos[sub]:
        datos[sub][run_num]['raw'] = preprocesar_raw(datos[sub][run_num]['raw'])
    print(f"{sub}: preprocesado ")

print("\nPreprocesamiento completo para todos los sujetos.")
```

```python
import numpy as np
import mne

# Parámetros de época
T_MIN, T_MAX = -1.0, 4.5
BASELINE     = (-1.0, 0.0)
# AMP_THRESHOLD = 100e-6  # LO COMENTAMOS PARA NO PERDER DATOS DEL SUB-003

# Mapeo evento → condición y código numérico
EVENTOS = {
    'TASK1T2': ('MR-D', 1), 'TASK1T1': ('MR-I', 2),
    'TASK2T2': ('IM-D', 3), 'TASK2T1': ('IM-I', 4),
    'TASK1T0': ('REST', 5), 'TASK2T0': ('REST', 5),
}

def segmentar_run(raw, df_events):
    sfreq   = raw.info['sfreq']

    # 1. CORRECCIÓN: Limpiar espacios invisibles en las etiquetas
    df_events['value'] = df_events['value'].astype(str).str.strip()

    df_filt = df_events[df_events['value'].isin(EVENTOS)].copy()
    if df_filt.empty:
        return {}

    samples  = (df_filt['onset'].values * sfreq).astype(int)
    ev_ids   = np.array([EVENTOS[v][1] for v in df_filt['value']])
    ev_array = np.column_stack([samples, np.zeros(len(samples), int), ev_ids])
    ev_dict  = {EVENTOS[v][0]: EVENTOS[v][1] for v in df_filt['value'].unique()}

    # 2. CORRECCIÓN: reject=None para evitar que descarte todo
    epochs = mne.Epochs(raw, ev_array, event_id=ev_dict,
                        tmin=T_MIN, tmax=T_MAX, baseline=BASELINE,
                        reject=None,  # <-- CLAVE PARA QUE NO DE 0 ÉPOCAS
                        preload=True, verbose=False)

    return {cond: epochs[cond] for cond in ev_dict
            if cond in epochs.event_id and len(epochs[cond]) > 0}

# Loop principal
epocas = {}
for sub in SUJETOS:
    print(f"\n{'='*40}\nSegmentando {sub}...")
    acumulado = {c: [] for c in ['MR-D','MR-I','IM-D','IM-I','REST']}

    for run_num, contenido in datos[sub].items():
        for cond, ep in segmentar_run(contenido['raw'], contenido['events']).items():
            if cond in acumulado:
                acumulado[cond].append(ep)
        print(f"  Run {run_num}: { {c: sum(len(e) for e in acumulado[c]) for c in acumulado} }")

    epocas[sub] = {c: mne.concatenate_epochs(l, verbose=False) if l else None
                   for c, l in acumulado.items()}

    for cond, ep in epocas[sub].items():
        n = len(ep) if ep is not None else 0
        print(f"  {'bien' if n>0 else 'mal'} {cond}: {n} épocas")

print("\n Segmentación completa.")
```
